```python
#無限級数の取り扱いがまだ

def lnfact(n):
    sum = 0
    for i in np.arange(1, n+1, 1):
        sum += np.log(i)
    return sum

def E_T(params,comission):
    a,b = params
    #print("a",a)
    #print("b",b)
    c= comission
    sum_list = []
    for i in range(1,40):
      k =2*i-1
      #dif = ((math.sqrt(2)*a)**k-(math.sqrt(2)*b)**k)/math.factorial(k)*math.gamma(k/2)
      if b>=0:
        dif = (math.exp(math.log(2*a*a)*(k/2))-math.exp(math.log(2*b*b)*(k/2)))/math.exp(lnfact(k))*math.gamma(k/2)
      else:
        dif = (math.exp(math.log(2*a*a)*(k/2))+math.exp(math.log(2*b*b)*(k/2)))/math.exp(lnfact(k))*math.gamma(k/2)
      sum_list.append(dif)
      #print(dif)
    sum_ = sum(sum_list)
    #print("beta:",beta)
    #print("alpha:",alpha)
    return_ = (a-b-c)/sum_


    return -return_
```
