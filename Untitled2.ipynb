```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pytorch_lightning as pl
from pytorch_lightning.callbacks import EarlyStopping
import torch

from pytorch_forecasting import Baseline, NHiTS, TimeSeriesDataSet
from pytorch_forecasting.data import NaNLabelEncoder
from pytorch_forecasting.data.examples import generate_ar_data
from pytorch_forecasting.metrics import SMAPE, MQF2DistributionLoss, QuantileLoss
```

```python
data = generate_ar_data(seasonality=10.0, timesteps=400, n_series=100, seed=42)
data["static"] = 2
data["date"] = pd.Timestamp("2020-01-01") + pd.to_timedelta(data.time_idx, "D")
data.head()
```

```python
max_encoder_length = 60
max_prediction_length = 20

training_cutoff = data["time_idx"].max() - max_prediction_length

context_length = max_encoder_length
prediction_length = max_prediction_length

training = TimeSeriesDataSet(
    data[lambda x: x.time_idx <= training_cutoff],
    time_idx="time_idx",
    target="value",
    categorical_encoders={"series": NaNLabelEncoder().fit(data.series)},
    group_ids=["series"],
    # only unknown variable is "value" - and N-HiTS can also not take any additional variables
    time_varying_unknown_reals=["value"],
    max_encoder_length=context_length,
    max_prediction_length=prediction_length,
)

validation = TimeSeriesDataSet.from_dataset(training, data, min_prediction_idx=training_cutoff + 1)
batch_size = 128
train_dataloader = training.to_dataloader(train=True, batch_size=batch_size, num_workers=0)
val_dataloader = validation.to_dataloader(train=False, batch_size=batch_size, num_workers=0)
```

```python
from datetime import datetime
import requests
import pandas as pd
import time

# 通貨ペアを指定
symbol = 'BTCUSDT'

# 時間足を指定(詳細はhttps://bybit-exchange.github.io/docs/linear/#kline-interval-interval)
interval = 5

# 取得開始の日付を指定
start_date_str = '2022/1/1 00:00'

# 取得終了の日付を指定
end_date_str = '2022/11/30 23:59'

# インバース無期限(https://bybit-exchange.github.io/docs/inverse/#t-querykline)
# url = 'https://api.bybit.com/v2/public/kline/list'

# USDT無期限(https://bybit-exchange.github.io/docs/linear/#t-querykline)
url = 'https://api.bybit.com/public/linear/kline'

timestamp = round(datetime.strptime(start_date_str, '%Y/%m/%d %H:%M').timestamp())
end_timestamp = round(datetime.strptime(end_date_str, '%Y/%m/%d %H:%M').timestamp())
klines = []

while timestamp < end_timestamp:
    params = {
        'symbol': symbol,
        'interval': interval,
        'from': timestamp,
        'limit': 200
    }

    response = requests.get(url, params=params)
    response_data = response.json()
    klines += response_data['result']
    timestamp += 200 * 60 * interval
    time.sleep(2)

data = pd.DataFrame(klines)

# 取得終了日付以降のデータを削除
data = data[data['open_time'] <= end_timestamp]

# タイムスタンプをdatetime型に変換
#data['open_time'] = data['open_time'].apply(datetime.fromtimestamp)

# 重複行を削除
data.drop_duplicates(subset=['open_time'], inplace=True)
data = data.reset_index(drop=True)

# 'open_time'カラムをインデックスに設定
#data.set_index('open_time', inplace=True)

# 不要なカラムを削除
data.drop(columns=['id', 'symbol', 'period', 'interval', 'start_at', 'turnover'], inplace=True)

data2 = data.copy()
data
```

```python
data2.loc[ : ,'before_close'] = data2["close"].shift(1)
#data2.loc[ : ,'before_high'] = data2["high"].shift(1)
#data2.loc[ : ,'before_low'] = data2["low"].shift(1)
#data2.loc[ : ,'before_volume'] = data2["volume"].shift(1)
data2.loc[ : ,'high_volatility'] = abs(data2["high"]/data2["before_close"]-1)*100
data2.loc[ : ,'low_volatility'] = abs(1-data2["low"]/data2["before_close"])*100
data2.loc[ : ,'high_low_volatility'] = abs(1-data2["high"]/data2["low"])*100
#data2.loc[ : ,'high'] = data2["high"]/1000
data2.loc[ : ,'series'] = 1
data2 = data2.drop('open', axis=1)
data2 = data2.drop('before_close', axis=1)
#data2 = data2.drop('close', axis=1)
#data2 = data2.drop('low', axis=1)
#data2 = data2.drop('volume', axis=1)
#data2 = data2.reset_index(inplace= True)

data2["open_time"] = data2["open_time"].apply(datetime.datetime.fromtimestamp)
#data2 = data2.reset_index(inplace= True)
data2=data2.dropna()
#data2 = data2.reset_index()
train_data = data2[:15000]
train_data = train_data.reset_index()
val_data = data2[15000:20000]
val_data = val_data.reset_index(drop=True)
val_data = val_data.reset_index()
pred_data = data2[20000:]
pred_data = pred_data.reset_index(drop=True)
pred_data = pred_data.reset_index()
train_data
```

```python
# create dataset and dataloaders
max_encoder_length = 50
max_prediction_length = 2

#training_cutoff = data["time_idx"].max() - max_prediction_length

context_length = max_encoder_length
prediction_length = max_prediction_length

training = TimeSeriesDataSet(
    train_data,
    time_idx="index",
    target="high_low_volatility",
    categorical_encoders={"series": NaNLabelEncoder().fit(train_data.series)},
    group_ids=["series"],
    # only unknown variable is "value" - and N-HiTS can also not take any additional variables
    time_varying_unknown_reals=["high_volatility","low_volatility","high","low","close","volume","high_low_volatility"],
    #time_varying_known_reals=["before_close","before_low","before_high","before_volume"],
    max_encoder_length=context_length,
    max_prediction_length=prediction_length,
)

validation = TimeSeriesDataSet.from_dataset(training, val_data)

batch_size = 128
train_dataloader = training.to_dataloader(train=True, batch_size=batch_size, num_workers=0)
val_dataloader = validation.to_dataloader(train=False, batch_size=batch_size, num_workers=0)
```

```python
actuals = torch.cat([y[0] for x, y in iter(val_dataloader)])
baseline_predictions = Baseline().predict(val_dataloader)
SMAPE()(baseline_predictions, actuals)
```

```python
pl.seed_everything(42)
trainer = pl.Trainer(gpus=0, gradient_clip_val=1.0)
net = NHiTS.from_dataset(
    training,
    learning_rate=3e-2,
    weight_decay=1e-2,
    loss=MQF2DistributionLoss(prediction_length=max_prediction_length),
    backcast_loss_ratio=0.0,
    hidden_size=64,
)
```

```python
res = trainer.tuner.lr_find(net, train_dataloaders=train_dataloader, val_dataloaders=val_dataloader, min_lr=1e-5)
print(f"suggested learning rate: {res.suggestion()}")
fig = res.plot(show=True, suggest=True)
fig.show()
net.hparams.learning_rate = res.suggestion()
```

```python
early_stop_callback = EarlyStopping(monitor="val_loss", min_delta=1e-4, patience=5, verbose=False, mode="min")
trainer = pl.Trainer(
    max_epochs=50,
    gpus=0,
    enable_model_summary=True,
    gradient_clip_val=1.0,
    callbacks=[early_stop_callback],
    #limit_train_batches=50,
    enable_checkpointing=True,
)


net = NHiTS.from_dataset(
    training,
    learning_rate=0.001,
    log_interval=10,
    log_val_interval=1,
    weight_decay=1e-2,
    backcast_loss_ratio=0.0,
    hidden_size=512,
    loss=MQF2DistributionLoss(prediction_length=max_prediction_length),
)

trainer.fit(
    net,
    train_dataloaders=train_dataloader,
    val_dataloaders=val_dataloader,
)
```

```python
best_model_path = trainer.checkpoint_callback.best_model_path
best_model = NHiTS.load_from_checkpoint(best_model_path)
```

```python
actuals = torch.cat([y[0] for x, y in iter(val_dataloader)])
predictions = best_model.predict(val_dataloader)
(actuals - predictions).abs().mean()
```
